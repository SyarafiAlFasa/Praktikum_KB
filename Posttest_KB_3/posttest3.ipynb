Import library pandas

```python
import pandas as pd
```

Import library matplotlib

```python
import matplotlib.pyplot as plt
```

Mengimport dataset dan memasukkan ke dalam variabel dataFrame

```python
dataFrame = pd.read_csv('Airlines.csv')
```

Melihat isi 5 daftar pertama

```python
dataFrame.head()
```

mengelompokkan data berdasarkan jenis Airline dan menghitung rata-rata dari setiap atributnya, kemudian memasukkannya kedalam variabel data

```python
data = dataFrame.groupby('Airline').mean()
data
```

Menampilkan grafik batang untuk menampilkan rata-rata jarak perjalanan berdasarkan jenis airlines. Terlihat bahwa CO memiliki rata-rata jarak perjalanan tertinggi dan YV memiliki rata-rata jarak perjalanan terendah

```python
x = data.index
y = data.Length
plt.bar(x, y, color=['green', 'blue', 'yellow'])
plt.title('Grafik Batang Rata-rata Jarak Perjalanan')
plt.xlabel('Jenis Maskapai')
plt.ylabel('Jarak Perjalanan')
plt.show()
```

Mengimport kembali dataset ke dalam variabel dataFrame

```python
dataFrame = pd.read_csv('Airlines.csv')
```

menghitung jumlah Delay pada dataset dan menampilkannya ke dalam Pie Chart dalam bentuk presentase. Terlihat bahwa presentase tepat waktu lebih tinggi dari pada presentase delay

```python
data = dataFrame.Delay.value_counts()
plt.pie(data, labels=data.index, autopct="%1.1f%%")
plt.title('Presentase Delay')
plt.legend()
plt.show()
```

Mengimport library seaborn

```python
import seaborn as sns
```

Menampilkan hubungan antar variabel dengan menggunakan heatmap. Terlihat hanya dua data yang mencolok karena memiliki angka yang rendah

```python
cor_matrix = dataFrame.corr()
sns.heatmap(cor_matrix, cmap='coolwarm', annot=True, center=0)
plt.title('Pemetaan Data')
plt.show()
```

Menghitung rata-rata berdasarkan jenis Airlines dan memasukkannya ke dalam variabel data

```python
data = dataFrame.groupby('Airline').mean()
data
```

Menampilkan sebaran data dengan box plot. Terlihat tidak terdapat data yang menjadi outliers

```python

sns.boxplot(x = data['Length'])
plt.title('Distribusi Data')
plt.show()
```
