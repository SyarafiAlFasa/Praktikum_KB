Data Split

```python
#import library
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
data = pd.read_csv('Airlines.csv')
data
```

```python
#subset target dan feature
X = data.iloc[:,:-1] #targer
y = data.iloc[:,-1] #feature
```

```python
# buat kelas train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3)
```

```python
print("Dimensi X_train : ", X_train.shape)
print("Dimensi X_train : ", X_test.shape)
print("Dimensi y_train : ", y_train.shape)
print("Dimensi y_test : ", y_test.shape)
```

Normalisasi Data

```python
from sklearn.preprocessing import MinMaxScaler

data_copy_min_max_scaler = data.copy()  # Salin dataset terlebih dahulu

# Pilih atribut yang ingin dinormalisasi, disini atribut yang dipilih adalah "Length"
attribute_to_normalize = 'Length'

# Inisialisasi Min-Max Scaler
scaler = MinMaxScaler()

# Ubah nilai atribut yang dipilih ke dalam bentuk matriks kolom dan lakukan normalisasi
data_copy_min_max_scaler[[attribute_to_normalize]] = scaler.fit_transform(data_copy_min_max_scaler[[attribute_to_normalize]])

# Tampilkan dataset setelah normalisasi
print(data_copy_min_max_scaler)
```

Standarisasi Data

```python
from sklearn.preprocessing import StandardScaler

data_copy_standard = data.copy()  # Salin dataset terlebih dahulu

# Inisialisasi Standard Scaler
scaler = StandardScaler()

# Pilih atribut yang ingin di-standarisasi 
attribute_to_standardize = 'Time'

# Standarisasi atribut yang dipilih
data_copy_standard[[attribute_to_standardize]] = scaler.fit_transform(data_copy_standard[[attribute_to_standardize]])

# Tampilkan dataset setelah standarisasi

print(data_copy_standard)
```

Data Cleaning

```python
#Pengecekan nilai null
print("Jumlah record yang memiliki nilai null:")
print(data.isna().sum())
```

```python
#Karena tidak ada nilai null maka kita akan mencoba membuat nilai null terlebih dahulu. Disini kita akan buat satu nilai null pada atribute DayOfWeek
data.loc[0, 'DayOfWeek'] = None
```

```python
#Pengecekan nilai null
print("Jumlah record yang memiliki nilai null:")
print(data.isna().sum())
```

```python
from sklearn.impute import SimpleImputer
```

```python
#Selanjutnya kita akan mengganti nilai null tersebut dengan modus
# Deklarasi kelas SimpleImputer dan lakukan transformasi pada kolom yang ingin diubah
imputer = SimpleImputer(strategy='most_frequent')
data["DayOfWeek"] = imputer.fit_transform(data[["DayOfWeek"]])
```

```python
# Melihat hasil dataframe yang berubah
data.head()
```

```python
#Melihat Jumlah Null sekarang
print("\n Jumlah null setelah menggunakan SimpleImputer: ")
print(data.isna().sum())
```

```python
#Selanjutanya kita akan cek nilai duplikat pada dataset
print("Nilai Duplikat: ")
data.duplicated().sum()
```

```python
#Kita lihat data head terlebih dahulu
data.head()
```

```python
#Karena tidak ada nilai duplikat maka kita akan membuatnya menjadi ada. Disini kita akan menggunakan metode .append()
#Menggunakan .append() untuk menambahkan data dengan indeks 0 (baris pertama) ke data untuk membuat duplikat
data = data.append(data.loc[0], ignore_index=True)
```

```python
#Sekarang, data pada indeks 0 telah menjadi duplikat dalam DataFrame.
data.head()
```

```python
#Selanjutanya kita akan cek nilai duplikat pada dataset
print("Nilai Duplikat: ")
data.duplicated().sum()
```

```python
#Kemudian Kita akan Hapus duplikat menggunakan metode drop_duplicates dari Pandas
data.drop_duplicates(inplace=True)
print("\nSetelah menggunakan fungsi drop_duplicates()")
print("Jumlah nilai Duplikat:", data.duplicated().sum())
```
